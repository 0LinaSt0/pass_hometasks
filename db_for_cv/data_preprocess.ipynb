## Unify

```python
# Let's unify the view, finding, date and location

import numpy as np


def unify_date_str(x):
    x_year = np.nan

    if isinstance(x, str):
        x_year = x[-4:]

        if '/' in x_year:
            x_year = '20' + x[-2:]
        elif ' ' in x_year:
            x_year = np.nan

    return x_year


df_metadata['finding'] = df_metadata['finding'].apply(
    lambda x: x.split('/')[-1].strip()
).apply(lambda x: 'Unknown' if x == 'todo' else x)

df_metadata['view'] = df_metadata['view'].apply(
    lambda x: 'APE' if x == 'AP Erect' else 'APS' if x == 'AP Supine' else x
)

df_metadata['date'] = df_metadata['date'].apply(unify_date_str)

df_metadata['location'] = df_metadata['location'].apply(
    lambda x: x.split(', ')[-1].strip() if isinstance(x, str) else x
)



display(df_metadata['finding'].drop_duplicates().sort_values().values)
display(df_metadata['view'].drop_duplicates().sort_values().values)
display(df_metadata['date'].drop_duplicates().sort_values().values)
display(df_metadata['location'].drop_duplicates().sort_values().values)
```

```python
import os
from pyspark.sql import SparkSession

def get_spark_session_hdfs(sparkname='COVID19_Analytics'):
    global spark
    try:
        spark
        spark_exists = True
    except NameError:
        spark_exists = False

    if spark_exists:
        spark.stop()

    spark = SparkSession.builder \
        .appName(sparkname) \
        .config('spark.hadoop.fs.defaultFS', 'hdfs://namenode:9000') \
        .config('spark.sql.warehouse.dir', 'hdfs://namenode:9000/user/jovyan/warehouse') \
        .enableHiveSupport() \
        .getOrCreate()
    return spark


def get_hdfs():
    global sc
    hadoop = sc._jvm.org.apache.hadoop
    conf = sc._jsc.hadoopConfiguration()
    conf.set('fs.defaultFS', 'hdfs://namenode:9000')
    return hadoop.fs.FileSystem.get(conf)

def hdfs_exists(path):
    fs = get_hdfs()
    return fs.exists(sc._jvm.org.apache.hadoop.fs.Path(path))


def hdfs_mkdir(path):
    fs = get_hdfs()
    p = sc._jvm.org.apache.hadoop.fs.Path(path)
    if not fs.exists(p):
        fs.mkdirs(p)
        print(f'Создана директория в HDFS: {path}')
    else:
        print(f'Директория в HDFS уже существует: {path}')



def hdfs_put(local_path, hdfs_path):
    fs = get_hdfs()
    src = sc._jvm.org.apache.hadoop.fs.Path(f'file://{local_path}')
    dst = sc._jvm.org.apache.hadoop.fs.Path(hdfs_path)
    fs.copyFromLocalFile(False, True, src, dst)
    print(f'Скопирован файл {local_path} в HDFS {hdfs_path}')


def count_files(path):
    return len([f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]) if os.path.exists(path) else 0
```

```python
import pyspark.sql.functions as F

df = spark.read.parquet('hdfs://namenode:9000/user/jovyan/covid_dataset/metadata/metadata_cleaned.parquet')

# Preprocess columns
df_prepared = df \
    .withColumn('patientid', F.col('patientid').cast('int')) \
    .withColumn('sex_filled', F.coalesce(F.col('sex'), F.lit('Unknown'))) \
    .withColumn('age_cleaned', F.col('age').cast('double')) \
    .withColumn('finding_unified', F.lower(F.regexp_replace(F.col('finding'), '\\s*/\\s*|\\s+', '_'))) \
    .select(
        'patientid',
        'filename',
        'age_cleaned',
        'sex_filled',
        'finding_unified',
        'clinical_notes'
    )

# Create bd
spark.sql('CREATE DATABASE IF NOT EXISTS covid_db')
spark.sql('USE covid_db')

spark.sql('DROP TABLE IF EXISTS covid_patients_optimized')
spark.sql('DROP TABLE IF EXISTS patient_outcomes')

# Create table
spark.sql('''
CREATE TABLE covid_patients_optimized (
    patientid INT,
    filename STRING,
    age_cleaned DOUBLE,
    sex_filled STRING,
    clinical_notes STRING
)
PARTITIONED BY (finding_unified STRING)
CLUSTERED BY (patientid) INTO 8 BUCKETS
STORED AS PARQUET
TBLPROPERTIES (
    'parquet.compression'='SNAPPY',
    'auto.purge'='true'
)
''')

# Create table for buckets
spark.sql('''
CREATE TABLE patient_outcomes (
    patientid INT,
    outcome STRING,
    days_hospitalized INT,
    severity STRING
)
CLUSTERED BY (patientid) INTO 8 BUCKETS
STORED AS PARQUET
''')

# Write data to tables
df_prepared.write \
    .mode('overwrite') \
    .partitionBy('finding_unified') \
    .bucketBy(8, 'patientid') \
    .sortBy('patientid') \
    .saveAsTable('covid_patients_optimized')

outcome_data = [
    (1, 'recovered', 10, 'mild'),
    (2, 'recovered', 14, 'moderate'),
    (3, 'deceased', 21, 'severe'),
    (4, 'recovered', 7, 'mild'),
    (5, 'deceased', 28, 'critical')
]

spark.createDataFrame(outcome_data, ['patientid', 'outcome', 'days_hospitalized', 'severity']) \
    .write.mode('overwrite') \
    .bucketBy(8, 'patientid') \
    .sortBy('patientid') \
    .saveAsTable('patient_outcomes')

# ANALITIC QUERIES
print('1. Оконная функция с партиционированием по диагнозу:')
spark.sql('''
SELECT 
    patientid,
    finding_unified,
    age_cleaned,
    AVG(age_cleaned) OVER (
        PARTITION BY finding_unified
        ORDER BY patientid
        ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING
    ) AS moving_avg_age,
    COUNT(*) OVER (PARTITION BY finding_unified) AS patients_in_diagnosis
FROM covid_patients_optimized
WHERE finding_unified IS NOT NULL
LIMIT 20
''').show(truncate=False)


print('\n2. JOIN между бакетированными таблицами:')
spark.sql('''
SELECT 
    c.patientid,
    c.finding_unified AS diagnosis,
    c.age_cleaned,
    p.outcome,
    p.days_hospitalized,
    p.severity,
    CASE 
        WHEN p.outcome = 'deceased' THEN 'high'
        WHEN c.age_cleaned > 60 THEN 'medium'
        ELSE 'low'
    END AS risk_level
FROM covid_patients_optimized c
JOIN patient_outcomes p ON c.patientid = p.patientid
ORDER BY p.days_hospitalized DESC
LIMIT 10
''').show(truncate=False)


print('\n3. Аналитический подзапрос с агрегацией:')
spark.sql('''
SELECT 
    c.finding_unified,
    COUNT(*) AS patients_count,
    ROUND(AVG(c.age_cleaned), 1) AS avg_age,
    (SELECT COUNT(*) FROM patient_outcomes WHERE outcome = 'deceased') AS total_deceased,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM covid_patients_optimized), 2) AS percentage
FROM covid_patients_optimized c
GROUP BY c.finding_unified
HAVING COUNT(*) > 5
ORDER BY patients_count DESC
LIMIT 10
''').show(truncate=False)
```

```python
from pyspark.sql.types import StringType
from pyspark.ml.feature import StringIndexer

# Create age cat
def cat_age(age):
    cat_age = 'unknown'
    if age:
        if age < 30:
            cat_age = 'young'
        elif age <= 60:
            cat_age = 'middle'
        else:
            cat_age = 'old'
    return cat_age


categorize_age_udf = F.udf(cat_age, StringType())
df_metadata = df_metadata.withColumn('age_cat', categorize_age_udf(F.col('age').cast('double')))

# Encode the sex
sex_indexer = StringIndexer(inputCol='sex_cat', outputCol='sex_index', handleInvalid='keep')
df_metadata = sex_indexer.fit(df_metadata).transform(df_metadata)

avg_age = df_metadata.select(F.avg(F.col('age').cast('double'))).first()[0]
df_metadata = df_metadata.withColumn(
    'age_filled', 
    F.when(F.col('age').isNull(), avg_age).otherwise(F.col('age').cast('double'))
)
```
