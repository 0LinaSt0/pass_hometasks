```python
def filework(fnc_exec, *args, **kwargs):
    ''' Function for working with file '''
    try:
        df = fnc_exec(*args, **kwargs)
        logging.info('Success work with file')
    except Exception as e:
        logging.error(f'During working with csv: {e}')
        df = None
    
    return df
```

# Tasks 12. Объединение двух DataFrame с использованием merge()

```python
!python -m venv tmp_venv
!source tmp_venv/bin/activate
!pip install pandas
!pip install logging
!pip freeze > tmp_requirements.txt
```

```python
import pandas as pd

import logging
```

```python
for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

logging.basicConfig(
    level=logging.INFO,
    format='%(levelname)s: %(message)s'
)
```

```python
def filework(fnc_exec, *args, **kwargs):
    ''' Function for working with file '''
    try:
        df = fnc_exec(*args, **kwargs)
        logging.info('Success work with file')
    except Exception as e:
        logging.error(f'During working with csv: {e}')
        df = None
    
    return df
```

```python
# Read file

df = filework(pd.read_csv, './example.csv')

df.head(5)
```

```python
# Create another df for merging

df['id'] = df.index

df_part = df.loc[
    :50, 
    ['race_ethnicity', 'math_score', 'reading_score', 'writing_score', 'total_score', 'id']
].copy()

df_part.shape
```

```python
# Merge dfs

df_merged = pd.merge(df, df_part, on='id')

df_merged.shape
```
