# Поиск карты глубины

Выводы по работе:
1. numDisparities отвечает за фон (чем меньше, тем больше белого): должен быть кратен 16
1. blockSize отвечает за зернистость: должен быть нечетным числом
1. флаговый параметр в cv2.imread отвечет за отображение:
    - (1) - cv2.IMREAD_COLOR - default - в RGB формате
    - (0) - cv2.IMREAD_GRAYSCALE - в графическом виде
1. в stereo.compute подается сначала левое, а потом правое изображение
1. правое изображение то, где на картинке больше виден правый бочок

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
MATERIALS_PATH = './materials/imgs/stereo_images/'
```

## Build

```python
picture = cv2.imread(MATERIALS_PATH + 'stereo_images1l.png', 1)
picture = cv2.cvtColor(picture, cv2.COLOR_BGR2RGB)

plt.imshow(picture)
plt.axis('off')  # Hide axes
plt.show()
```

```python
img_r = cv2.imread(MATERIALS_PATH + 'stereo_images1r.png', 0)
img_l = cv2.imread(MATERIALS_PATH + 'stereo_images1l.png', 0)
```

```python
stereo = cv2.StereoBM_create(numDisparities=48, blockSize=9)
disparity = stereo.compute(img_l, img_r)

plt.figure(figsize= (10, 5))
plt.imshow(disparity, 'gray')
plt.xticks()
plt.yticks()
plt.show()
```
