# Task 8 & 12. Две линии на одном графике с легендой. График с второстепенной осью Y (twinx)

```python
!python -m venv tmp_venv
!source tmp_venv/bin/activate
!pip install numpy
!pip install matplotlib
!pip freeze > tmp_requirements.txt
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
time = np.arange(0, 10, 0.5)  # Time from 0 to 10 seconds
temperature = np.random.uniform(20, 30, size=len(time))  # Random temperatures between 20 and 30 degrees Celsius
humidity = np.random.uniform(30, 90, size=len(time))  # Random humidity between 30% and 90%

# Create the figure and the first axis
fig, ax1 = plt.subplots()

# Plot temperature on the first y-axis
ax1.set_xlabel('Time, sec')
ax1.set_ylabel('Temperature, C', color='tab:red')
l1, = ax1.plot(time, temperature, color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity, %', color='tab:blue')
l2, = ax2.plot(time, humidity, color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Add a title and show the plot
plt.title('Temperature and humidity')

# Add legend
lines = [l1, l2]
labels = [l.get_label() for l in lines]
plt.legend(lines, labels)

plt.show()
```
